## EDA & Visualizations


##  1-(Salary Analysis)


- 1-Distribution of Median Salary
- - Top 10 Highest Paying Majors
- -  - 3-Top 10 Lowest Paying Majors
-  - - - 4-Average Salary by Major Category

####  --1-Distribution of Median Salary


```python
fig = px.histogram(
    df,
    x="Median",
    nbins=25,
    title="Distribution of Median Salaries",
    color_discrete_sequence=["#00CC96"],
    template="plotly_dark"
)
fig.add_vline(
    x=df["Median"].mean(),
    line_dash="dash",
    line_color="red",
    annotation_text="Mean"
)

fig.add_vline(
    x=df["Median"].median(),
    line_dash="dot",
    line_color="yellow",
    annotation_text="Median"
)

fig.update_layout(
    title_x=0.5,
    xaxis_title="Median Salary",
    yaxis_title="Number of Majors",
    bargap=0.05
)
fig.update_xaxes(
    tickprefix="$",
    separatethousands=True
)

fig.show()
```


The mean salary is higher than the median salary, indicating a right-skewed distribution caused by a few high-paying majors.

#### --2-Top 10 Highest Paying Majors


```python
top_10_highest=df.head(10)
```

```python
fig =px.bar(top_10_highest.sort_values(by="Median",ascending=True),
            x='Median',y="Major",
            title="Top 10 Highest Paying College Majors",
            color="Median",
            text="Median",
            color_continuous_scale="Viridis",
            orientation="h",
            template="plotly_dark")
fig.update_layout(
    title_x=0.5,
    xaxis_title="Median Salary",
    yaxis_title="Major"
)
category_orders={
    "Major": top_10_highest["Major"][::-1]
}

fig.update_traces(
    textposition="outside",
    texttemplate="$%{x:,.0f}"
)

fig.show()
```

#### 3-Top 10 Lowest Paying Majors


```python
top_low = df.tail(10)
```

```python
fig = px.bar(top_low.sort_values(by="Median",ascending=True),
             x="Median",y="Major",
             color="Median",color_continuous_scale="Viridis",
             template="plotly_dark",
             text="Median",
             orientation="h",
             title="Top 10 lowest Paying College Majors"
)
fig.update_layout(
    title_x=0.5,
    xaxis_title="Median",
    yaxis_title="Major"
)
category_orders={
    "Major": top_low["Major"][::-1]
}
fig.update_traces(
    textposition="outside",
    texttemplate="$%{x:,.0f}"
)
fig.show()
```

#### 4-Average Salary by Major Category

```python
category_salary=(df.groupby("Major_category",as_index=False)["Median"].mean())
```

```python
fig = px.bar(
    category_salary.sort_values(by="Median",ascending=False),
    x="Major_category",
    y="Median",
    color="Median",
    text="Median",
    template="plotly_dark",
    title="Average Salary by Major Category",
    orientation="v"
)

fig.update_layout(
    title_x=0.5,
    xaxis_title="Major Category",
    yaxis_title="Average Median Salary ($)",
    xaxis_tickangle=-45
)
fig.update_traces(
    textposition="outside",
    texttemplate="$%{y:,.0f}"

)
fig.show()
```

## 2-Employment Analysis

- Top 10 Highest Unemployment Rate
- - Top 10 Lowest Unemployment Rate
- - - Salary vs Unemployment Rate (Scatter Plot)

#### 1-Top 10 Highest Unemployment Rate


```python
top_10_high=df.nlargest(10,"Unemployment_rate").sort_values('Unemployment_rate')
```

```python
plt.figure(figsize=(12,6))

ax = sns.barplot(
    data=top_10_high,
    x="Unemployment_rate",
    y="Major",
palette="Reds_r"
)
ax.xaxis.set_major_formatter(PercentFormatter(xmax=1))

plt.title("Top 10 Highest Unemployment Rate", fontsize=16)
plt.xlabel("Unemployment Rate")
plt.ylabel("Major")
plt.style.use("dark_background")
plt.show()
```

#### 2- Top 10 Lowest Unemployment Rate


```python
top_10_lowest=df.nsmallest(10,"Unemployment_rate").sort_values('Unemployment_rate')
```

```python
plt.figure(figsize=(12,6))

ax = sns.barplot(
    data=top_10_lowest,
    x="Unemployment_rate",
    y="Major",
palette="Reds_r"
)
ax.xaxis.set_major_formatter(PercentFormatter(xmax=1))

plt.title("Top 10 Lowest  Unemployment Rate", fontsize=16)
plt.xlabel("Unemployment Rate")
plt.ylabel("Major")
plt.style.use("dark_background")
plt.show()
```

#### 3-Salary vs Unemployment Rate (Scatter Plot)

```python
plt.figure(figsize=(12,7))

sns.scatterplot(
    data=df,
    x="Median",
    y="Unemployment_rate",
    hue="Major_category",
    s=100
)

plt.title("Salary vs Unemployment Rate by Major Category")
plt.xlabel("Median Salary ($)")
plt.ylabel("Unemployment Rate")

plt.legend(bbox_to_anchor=(1.02,1), loc="upper left")

plt.show()
```

## 3-(Gender Analysis)

- Distribution of ShareWomen
- Top Female-Dominated Majors
- Top Male-Dominated Majors


#### 1- Distribution of ShareWomen


```python
plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x="ShareWomen",
    bins=15,
    kde=True,
    color="skyblue"
)

plt.title("Distribution of ShareWomen")
plt.xlabel("Share of Women")
plt.ylabel("Number of Majors")

plt.show()
```

#### 2- Top Female-Dominated Majors


```python
top_female = (
    df.nlargest(10, "ShareWomen")
      .sort_values("ShareWomen")
)
```

```python
plt.figure(figsize=(12,6))

ax = sns.barplot(
    data=top_female,
    x="ShareWomen",
    y="Major",
    palette="Blues_r"
)

for container in ax.containers:
    ax.bar_label(container, fmt="%.2f")

plt.title("Top 10 Female-Dominated Majors", fontsize=16)
plt.xlabel("Share of Women")
plt.ylabel("Major")

plt.show()
```

#### 3- Top Male-Dominated Majors


```python
top_male = (
    df.nsmallest(10, "ShareWomen")
      .sort_values("ShareWomen")
)
```

```python
plt.figure(figsize=(12,6))

ax = sns.barplot(
    data=top_male,
    x="ShareWomen",
    y="Major",
    palette="Reds_r"
)

ax.xaxis.set_major_formatter(PercentFormatter(xmax=1))

for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%")

plt.title("Top 10 Male-Dominated Majors", fontsize=16)
plt.xlabel("Share of Women (%)")
plt.ylabel("Major")

plt.show()
```
